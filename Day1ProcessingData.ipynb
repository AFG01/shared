```python
data = np.loadtxt('SdH1.dat')
B = data[:, 0]
Rxx = data[:, 3]
```

```python
plt.plot(1/B[3000:], Rxx[3000:])
```

```python
import scipy.interpolate

f = scipy.interpolate.interp1d(1/B[3000:], Rxx[3000:])
xvalues = np.linspace(1.7,3.6,10*len(1/B[3000:]))
```

```python
plt.plot(f(xvalues))
```

```python
xvalues
```

```python
Z = np.fft.fft(f(xvalues))

freq = np.fft.fftfreq(len(xvalues), np.abs(xvalues[1]-xvalues[0]))
sfreq = np.fft.fftshift(freq)

plt.plot(sfreq, np.abs(Z))

plt.ylim([0., 10000])
plt.grid()
```
